```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Customer 360+ Final_Project/Dataset/Cleaned_dataset.csv")
```

```python
df.head(5)
```

### Compare Churned vs Non-Churned

```python
metrics = ["TotalRevenue", "TotalOrders", "TotalVisits", "SupportTickets", "SatisfactionScore"]

for metric in metrics:
    plt.figure(figsize=(7,5))
    sns.barplot(
        data=df,
        x="Churned", y=metric,
        estimator="mean",
        palette={"Yes": "Purple", "No": "Red"}, #tomato,seagreen
        ci="sd"
    )
    plt.title(f"Average {metric} by Churned Status", fontsize=14, weight="bold")
    plt.xlabel("Churned", fontsize=12)
    plt.ylabel(f"Avg {metric}", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.savefig(f"D:/Customer 360+ Final_Project/Notebooks/03_Diagnostic_Analysis/Output/{metric}_by_churned.png", dpi=300, bbox_inches="tight")
    plt.show()
```

### Correlation Matrix

```python
plt.figure(figsize=(10,6))
corr = df[metrics].corr()

sns.heatmap(
    corr,
    annot=True, fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    square=True,
    linewidths=0.5,
    annot_kws={"size":10}
)
plt.title("Correlation Matrix of Numerical Variables", fontsize=14, weight="bold")
plt.savefig("D:/Customer 360+ Final_Project/Notebooks/03_Diagnostic_Analysis/Output/Correlation_matrix.png")
plt.show()
```

### Boxplots / Violin plots to check distributions & outliers

```python
for metric in metrics:
    plt.figure(figsize=(7,5))
    sns.boxplot(
        data=df,
        x="Churned", y=metric,
        palette={"Yes": "tomato", "No": "seagreen"}
    )
    plt.title(f"Boxplot of {metric} by Churned Status", fontsize=14, weight="bold")
    plt.xlabel("Churned", fontsize=12)
    plt.ylabel(metric, fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.savefig(f"D:/Customer 360+ Final_Project/Notebooks/03_Diagnostic_Analysis/Output/Boxplot_{metric}_by_churned.png", dpi=300, bbox_inches="tight")
    plt.show()

    
    plt.figure(figsize=(7,5))
    sns.violinplot(
        data=df,
        x="Churned", y=metric,
        palette={"Yes": "tomato", "No": "seagreen"},
        inner="quartile"
    )
    plt.title(f"Violin Plot of {metric} by Churned Status", fontsize=14, weight="bold")
    plt.xlabel("Churned", fontsize=12)
    plt.ylabel(metric, fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.savefig(f"D:/Customer 360+ Final_Project/Notebooks/03_Diagnostic_Analysis/Output/Violinplot_{metric}_by_churned.png", dpi=300, bbox_inches="tight")
    plt.show()
```
